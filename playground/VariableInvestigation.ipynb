```python
import sqlite3
import numpy as np
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
%matplotlib notebook
%config InlineBackend.figure_format = 'retina'
```

```python
conn = sqlite3.connect("minion_1016_sqlite.db")
cur = conn.cursor()
```

```python
cur.execute("""select FWHMeff from summary""")
seeing = np.asarray(cur.fetchall())[:,0]
```

```python
cur.execute("""select airmass from summary""")
airmass = np.asarray(cur.fetchall())[:,0]
```

```python
fig, ax = plt.subplots()

ax.hist(seeing, bins = 100, alpha = 0.5, label = 'seeing')
ax.hist(airmass, bins = 100, alpha = 0.5, label = 'airmass')
ax.legend(loc = 'upper right')

fig.tight_layout()
```

```python
fig, ax = plt.subplots()

ax.hist(seeing, bins = 100, alpha = 0.5, label = 'seeing')
ax.hist(airmass**0.6, bins = 100, alpha = 0.5, label = 'airmass ^ 0.6')
ax.legend(loc = 'upper right')

fig.tight_layout()
```
